# Seattle Landmarks

```python
import pandas as pd
```

```python
data = pd.read_csv("Landmarks.csv", names=('City Feature', 'Common Name', 'Address', 'Website', 'Latitude', 'Longitude'))
data
```

```python
website_data = data[pd.notnull(data['Website'])]
```

## What percentage of the landmarks have websites?

```python
'{:0.2f}% of the landmarks have websites.'.format((len(website_data)/len(data)) * 100)
```

```python
school_data = data[data['Common Name'].str.contains("School")]
```

## How many schools are considered Landmarks?

```python
'{} of the landmarks are schools.'.format(len(school_data))
```

## How evenly dispersed are the Landmarked schools?

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(school_data["Latitude"], school_data["Longitude"], c="blue", label="unbound by Charles River")
plt.xlabel("Latitude"), plt.ylabel("Longitude"), plt.title("The dispersal of landmarked schools.")
plt.show()
```

## How close are the landmarked libraries to the landmarked schools?

```python
library_data = data[data['Common Name'].str.contains("Library")]
```

```python
plt.scatter(school_data["Latitude"], school_data["Longitude"], c="blue", label="Schools")
plt.scatter(library_data["Latitude"], library_data["Longitude"], c="red", label="Libraries")
plt.xlabel("Latitude"), plt.ylabel("Longitude"), plt.title("How close are the schools to the libraries")
plt.legend(loc='upper left')
plt.show()
```

## How many landmarks are around the space needle?

```python
space_needle = data[data['Common Name'].str.contains("Space Needle")]
data = data[data["Common Name"] != "Common Name"]
```

```python
plt.scatter(data["Latitude"], data["Longitude"])
plt.scatter(space_needle["Latitude"], space_needle["Longitude"], c="red", label="Space Needle")
plt.xlabel("Latitude"), plt.ylabel("Longitude"), plt.title("How many other landmarks are around the space needle?")
plt.legend(loc='upper left')
plt.show()
```
