```python
data = pd.read_csv("boston_housing_data.csv", delimiter="\s+", names=('CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX', 'PTRATIO', 'B', 'LSTAT', 'MEDV'))
data
```

The median and average home values across all Boston Suburbs in dollars.

```python
data = pd.DataFrame(data)
```

```python
'Median Home Value: ${:0.2f}'.format(data["MEDV"].median() * 1000)
```

```python
'Average Home Value: ${:0.2f}'.format(data["MEDV"].mean() * 1000)
```

The median home value of the suburb with the newest houses.

```python
newest = data.loc[data['AGE'].idxmin()]
'Median home value of suburb with newest houses: {:0.2f}'.format(newest['MEDV'] * 1000)
```

The relationship between per-capita crime rate and the pupil-teacher ratio. Differentiate between whether or not the suburb is bounded by the Charles River.


```python
unbounded = data[data["CHAS"]==0].filter(["CRIM", "PTRATIO"])
```

```python
bounded = data[data["CHAS"]==1].filter(["CRIM", "PTRATIO"])
```

```python
plt.scatter(unbounded["PTRATIO"], unbounded["CRIM"], c="blue", label="unbound by Charles River")
plt.scatter(bounded["PTRATIO"], bounded["CRIM"], c="red", label="bound by Charles River")
plt.xlabel("Crime"), plt.ylabel("Pupil-Teacher Ratio"), plt.title("Crime rate versus Pupil Teacher Ratio")
plt.legend(loc='upper left')
plt.show()
```

The relationship between the proportion of black citizens and the distance to employment centers in Boston.


```python
perc_black_dis_to_employ_ctr = data.filter(["B","DIS"])
plt.scatter(perc_black_dis_to_employ_ctr.B, perc_black_dis_to_employ_ctr.DIS, c='red')
plt.xlabel("Proportion of Black People"), plt.ylabel("Distance to Employment Center")
plt.title("Proportion of Black People vs. Weighted Distance to Employment Center")
plt.show()
```

The relationship between median value of owner-occuped homes and nitric oxide concentration along with median home value and the proportion of non-retail business (on the same plot).

```python
med_home_val_nox = data.filter(["MEDV","NOX"])
med_home_val_indus = data.filter(["MEDV", "INDUS"])
plt.scatter(med_home_val_nox.MEDV, med_home_val_nox.NOX, c='red')
plt.scatter(med_home_val_indus.MEDV, med_home_val_indus.INDUS, c='blue')
plt.xlabel("Median Value of Homes")
plt.ylabel("NOX Concentration, Proportion of Non-Retail Business")
plt.show()
```

Q: How does the crime rate relate to the median home value?
A: As expected, the crime rate is much higher where the median home value is lower.

```python
med_home_value_and_crime = data.filter(["MEDV","CRIM"])
plt.scatter(med_home_value_and_crime.CRIM, med_home_value_and_crime.MEDV, c='red')
plt.xlabel("Per Capita Crime rate"), plt.ylabel("Median Home Value")
plt.title("How does the crime rate compare the median home value?")
plt.show()
```
